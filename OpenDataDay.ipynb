```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import json

# Paso 1: Cargar el archivo CSV de IPRESS
df2 = pd.read_csv('RENIPRESS_2025_v2.csv',delimiter=';', encoding='ISO-8859-1')

# Asegurarse de que los nombres de los departamentos estÃ©n en mayÃºsculas
df2['DEPARTAMENTO'] = df2['DEPARTAMENTO'].str.upper()

# Contar la cantidad de IPRESS por departamento
ipress_por_departamento = df2['DEPARTAMENTO'].value_counts().reset_index()
ipress_por_departamento.columns = ['DEPARTAMENTO', 'IPRESS_COUNT']

# Paso 2: Cargar el archivo GeoJSON de los departamentos de PerÃº
geojson_path = 'peru_departamental_simple.geojson'
gdf = gpd.read_file(geojson_path)

# Paso 3: Fusionar el GeoDataFrame con el DataFrame de IPRESS
gdf_merged = gdf.merge(ipress_por_departamento, how='left', left_on='NOMBDEP', right_on='DEPARTAMENTO')

# Paso 4: Crear el mapa de calor
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf_merged.plot(column='IPRESS_COUNT', ax=ax, legend=True,
                legend_kwds={'label': "Cantidad de IPRESS por Departamento",
                             'orientation': "horizontal"},
                cmap='YlOrRd')

# Agregar tÃ­tulo y mostrar el grÃ¡fico
plt.title("Mapa de Calor de IPRESS por Departamento")
plt.show()
```
